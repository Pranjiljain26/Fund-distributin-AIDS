```python
df['Text'].to_csv('text added.csv', index=False)
```

```python
df_text=pd.read_csv("text added.csv")
```

```python
df_text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df_text['Text'])

# Step 3: Clustering
num_clusters = 5  # Number of desired clusters
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)

# Step 4: Add cluster labels to DataFrame
df_text['Cluster'] = kmeans.labels_

# Step 5: Print DataFrame with cluster labels
df_text.head()
```

```python
unique_names = df_text['Cluster'].unique()
```

```python
unique_names
```

```python
mapping = {0: 'Very low funds required',
           1: 'low funds',
           2: 'normal',
           3: 'high funds',
          4:"very high requirment of funds"}

# Using replace() method
df_text['Cluster'] = df_text['Cluster'].replace(mapping)
```
